# Exploratory Data Analysis (SQL)
Performing data analysis using SQL queries to extract specific insights about the SpaceX dataset.

```python
import sqlite3
import pandas as pd

# 1. Load the dataset
# Ensure the path is correct relative to your notebook location
df = pd.read_csv('../data/dataset_part_2.csv')

# 2. Clean column names (Optional but recommended)
# This removes any hidden spaces that might cause SQL errors
df.columns = df.columns.str.strip()

# 3. Create the in-memory SQL database
conn = sqlite3.connect(':memory:')
df.to_sql('SPACEXTABLE', conn, index=False, if_exists='replace')

# --- QUERIES ---

# Query 1: Unique Launch Sites
query1 = "SELECT DISTINCT LaunchSite FROM SPACEXTABLE"
print("--- Unique Launch Sites ---")
print(pd.read_sql(query1, conn))
print("\n")

# Query 2: CCAFS launches (First 5)
query2 = "SELECT * FROM SPACEXTABLE WHERE LaunchSite LIKE 'CCA%' LIMIT 5"
print("--- CCAFS Launch Records ---")
print(pd.read_sql(query2, conn))
print("\n")

# Query 3: Total Payload Mass for NASA (CRS)
# Corrected 'PayloadMassKG' to 'PayloadMass' to match your dataframe
# Note: If this returns 0 or None, verify 'Customer' exists in your CSV
try:
    query3 = "SELECT SUM(PayloadMass) AS Total_Payload_Mass FROM SPACEXTABLE WHERE Customer = 'NASA (CRS)'"
    print("--- Total Payload Mass (NASA CRS) ---")
    print(pd.read_sql(query3, conn))
except Exception as e:
    print(f"Query 3 failed: {e}")
    print("Available columns are:", df.columns.tolist())
```
