# Exploratory Data Analysis (Visualization)
In this notebook, we visualize the relationships between flight numbers, launch sites, payload mass, and landing success.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the wrangled data
df = pd.read_csv('../data/dataset_part_2.csv')

# Plotting Flight Number vs. Launch Site
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect=5)
plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Launch Site", fontsize=20)
plt.show()

# Success Rate by Orbit Type
orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()
sns.barplot(x="Orbit", y="Class", data=orbit_success)
plt.title("Success Rate by Orbit Type")
plt.show()
```
