# SpaceX Data Collection
In this notebook, we use the SpaceX API to retrieve historical launch data.

```python
import requests
import pandas as pd
import numpy as np

# API URL for SpaceX Launches
url = "https://api.spacexdata.com/v4/launches/past"
response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.json_normalize(data)

# Selecting relevant columns
df = df[['static_fire_date_utc', 'rocket', 'payloads', 'success', 'launchpad', 'flight_number', 'date_utc']]
df.head()
```
