# Interactive Visual Analytics with Folium and Dash
In this notebook, we visualize launch site locations on interactive maps and build a dashboard to analyze launch success rates based on payload and site.

```python
# Run this if you haven't installed these yet
%pip install folium plotly dash 

import pandas as pd
import folium
import plotly.express as px
from dash import Dash, dcc, html

# Load the data - make sure this runs successfully!
df = pd.read_csv('../data/dataset_part_2.csv')

print("Data loaded successfully. Columns found:", df.columns.tolist())
```

```python
# Initialize map at Kennedy Space Center
site_map = folium.Map(location=[28.57, -80.64], zoom_start=5)

# We use unique launch sites to avoid plotting 90 markers on top of each other
unique_sites = df.groupby('LaunchSite')[['Latitude', 'Longitude', 'Class']].mean().reset_index()

for index, row in unique_sites.iterrows():
    color = 'green' if row['Class'] > 0.5 else 'red' # Green if success rate > 50%
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        icon=folium.Icon(color=color),
        popup=f"{row['LaunchSite']} (Avg Success: {row['Class']:.2f})"
    ).add_to(site_map)

site_map
```

```python
# Success by Payload Mass scatter plot
# Ensure x matches your column name (PayloadMass)
fig = px.scatter(df, x="PayloadMass", y="Class", color="BoosterVersion",
                 title="Payload vs. Outcome (1=Success, 0=Failure)")

fig.show()
```
