# Data Wrangling and Preprocessing
In this section, we clean the dataset and prepare the 'Class' column (1 for success, 0 for failure).

```python
import pandas as pd
import numpy as np
import os

# 1. Load the data
file_path = '../data/dataset_part_1.csv'
print(f"Checking for file at: {file_path}")

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("✅ Success: dataset_part_1.csv loaded.")

    # 2. Identify the landing outcomes
    landing_outcomes = df['Outcome'].value_counts()
    print("\n--- Landing Outcomes Counts ---")
    print(landing_outcomes)

    # 3. Create the 'Class' column (The Core Logic)
    # We assign 1 to successful landings and 0 to unsuccessful ones.
    # Success strings usually start with 'True' (e.g., True ASDS, True RTLS)
    df['Class'] = df['Outcome'].apply(lambda x: 1 if 'True' in str(x) else 0)

    # 4. Verification of the new column
    print("\n--- Success Rate (Class 1 vs 0) ---")
    print(df['Class'].value_counts())
    
    # Calculate and show the mean success rate
    success_rate = df['Class'].mean()
    print(f"\nOverall Success Rate: {success_rate:.2%}")

    # 5. Export the wrangled data to Part 2
    output_path = '../data/dataset_part_2.csv'
    df.to_csv(output_path, index=False)
    
    print(f"\n✅ PROCESS COMPLETE: Wrangled data saved to {output_path}")
    
    # Show the first 5 rows to confirm columns
    display(df.head())
else:
    print(f"❌ ERROR: File not found at {file_path}.")
    print("Ensure your notebook is in the 'notebooks' folder and the data is in the 'data' folder.")
```
